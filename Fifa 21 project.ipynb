```python
import pandas as pd
```

```python
##extract Fifa21 file in pandas
```

```python
df = pd.read_csv('Fifa21.csv')
```

```python
#remove columns from nation_jersey_number
```

```python
df.drop(df.columns[32:], axis = 1, inplace = True)
```

```python
#remove colum with picture link player_url
```

```python
df.drop(columns = 'player_url', inplace = True)
```

```python
#remove any column or row with NAN
```

```python
df.dropna(how = 'any', axis = 1, inplace = True)
```

```python
#remove colum with names not specified 
```

```python
df = df[~df['body_type'].str.contains('PLAYER_BODY_TYPE', na=False)]
```

```python
#import matplotlib & seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##what is the top 10 player by overall rating ditributed across the dataset
```

```python
top_10 = df.sort_values('overall', ascending = False)[['short_name', 'overall', 'age']].head(10)
```

```python
top_10
```

```python
sns.barplot(data = top_10, x= 'short_name', y= 'overall', hue= 'age', width = 0.7)
plt.title('players vs ratings')
plt.xlabel('players')
plt.ylabel('overall ratings')
```

```python
# What is the average market value of players in different positions?
```

```python
market= df.sort_values('overall', ascending = False).head(10)
```

```python
average_market = market.groupby(['short_name', 'overall', 'player_positions'])['value_eur'].mean()
```

```python
#Which positions have the highest average market value?
```

```python
average_market_highest = df.groupby(['short_name','player_positions'])['value_eur'].mean().nlargest(1)
```

```python
average_market_highest
```

```python
#Which positions have the lowest average market value?
```

```python
average_market_lowest = df.groupby(['short_name','player_positions'])['value_eur'].mean().nsmallest(1)
```

```python
average_market_lowest
```
